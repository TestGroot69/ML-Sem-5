```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
d=pd.read_csv("D://Code/ML Sem 5/Datasets/temperatures.csv")
d
```

```python
d.shape
```

```python
d.head()
```

```python
x_training=d[["YEAR"]]
y_training=d[["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]]
```

```python
model=LinearRegression()
```

```python
model.fit(x_training,y_training)
```

```python
predicted=model.predict(x_training)
```

```python
predicted
```

```python
mean_squared_error(y_training,predicted)
```

```python
mean_absolute_error(y_training,predicted)
```

```python
r2_score(y_training,predicted)
```

```python
columns_to_plot = [col for col in d.columns if col != 'YEAR']
for column in columns_to_plot:
    plt.scatter(d['YEAR'], d[column], label=column)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.title('Scatter Plots of Month-Wise Temperature against year')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
correlation=d.corr()["YEAR"].sort_values(ascending=False)
```

```python
correlation
```

```python
#Plotting graph between HIGHLY correlated features and year

plt.plot(d["YEAR"], d["DEC"])
plt.xlabel("Year")
plt.ylabel('Dec')
plt.title('Graph')
plt.show()
```

```python
x=d[["YEAR"]]
y=d[["DEC"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
model.fit(x_train,y_train)
```

```python
y_test_lr =model.predict(x_test)
y_train_lr =model.predict(x_train)
```

```python
print('R-Squared Error :',r2_score(y_test,y_test_lr))
print('Mean Absolute Error :',mean_absolute_error(y_test,y_test_lr))
print('Mean Squared Error :',mean_squared_error(y_test,y_test_lr))
print('Root Mean Squared Error :',np.sqrt(mean_squared_error(y_test,y_test_lr)))
```

```python
plt.scatter(x_test,y_test,color = 'green');
plt.scatter(x_train,y_train,color = 'orange');
plt.plot(x_train,y_train_lr, color = 'black');
plt.legend(['Testing Set','Training Set','Best fit Regression line'])
plt.title('Temperature vs Year for month Dec')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show();
```

```python
#Plotting graph between LEAST correlated feature and year

plt.plot(d["YEAR"], d["JUN"])
plt.xlabel("Year")
plt.ylabel('JUN')
plt.title('Graph')
plt.show()
```

```python
x=d[["YEAR"]]
y=d["JUN"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
model.fit(x_train,y_train)

y_test_lr = model.predict(x_test)
y_train_lr = model.predict(x_train)
```

```python
print('R-Squared Error :',r2_score(y_test,y_test_lr))
print('Mean Absolute Error :',mean_absolute_error(y_test,y_test_lr))
print('Mean Squared Error :',mean_squared_error(y_test,y_test_lr))
print('Root Mean Squared Error :',np.sqrt(mean_squared_error(y_test,y_test_lr)))
```

```python
plt.scatter(x_test,y_test,color = 'green');
plt.scatter(x_train,y_train,color = 'orange');
plt.plot(x_train,y_train_lr, color = 'black');
plt.legend(['Testing Set','Training Set','Best fit Regression line'])
plt.title('Temperature vs Year for month Jun')
plt.ylabel('Temperature')
plt.xlabel('Year')
plt.show();
```
